```python
#Import Dependencies
import pandas as pd
import numpy as np
housingDF = pd.read_csv("../Resources/housingDataUpdated.csv")
```

```python
#Neighborhood has 662 unknowns
housingDF.groupby("neighborhood").nunique()
```

```python
#lot_size has 692 NaN values
housingDF['lot_size'].isna().sum()
```

```python
#dropping extraneous columns
allCleanHousing = housingDF.drop(columns={"address", "city", "lot_size", "neighborhood", 
                                          "middle_school", "elementary_school"})
```

```python
allCleanHousing.nunique()
```

```python
#dropping high_school where string == Current Price:
counter = 1
for index, row in allCleanHousing.iterrows():
    if (row['high_school'] == "Current Price:") or (row['high_school'] == "Other"):
        allCleanHousing.drop(index, inplace=True)
        print(f"removed: {counter} row(s)")
        counter +=1
```
